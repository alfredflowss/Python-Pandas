```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

import matplotlib.pyplot as plt
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#grouping genders
uniquegender = purchase_data.groupby('Gender')['SN'].nunique()
gender = pd.DataFrame(uniquegender)

#calculating values for percentages for each gender 
gender_sum = gender['SN'].sum()

gender_percent = round(((gender["SN"]/gender_sum) * 100), 2)

#putting my calculated values into a df
gender["Percentage Of Players"] = gender_percent.astype(str) + "%"


gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#grouping genders
groupgender = purchase_data.groupby(["Gender"])
groupgender["Purchase ID"].count()

#calculating my values for avg. purchase price, total purchase value, avg. purchase total per person 
mean_price =  round(groupgender["Price"].mean(), 2)
total_purchase = groupgender["Price"].sum()
avg_tprice = round(total_purchase/gender["SN"], 2)

avg_tprice = pd.DataFrame(avg_tprice)

#merging results based on Gender which is the primary key
merger = pd.merge(groupgender["Purchase ID"].count(), mean_price, on ="Gender")
merger = pd.merge(merger, total_purchase, on ="Gender")
merger = pd.merge(merger, avg_tprice, on ="Gender")

#cleaninng up my data and putting them in a summary df
analysed_gender = merger.rename(columns={'Purchase ID':"Purchase Count", 'Price_x':"Average Purchase Price", 
                                 'Price_y':"Total Purchase Value", 0:"Avg Total Purchase per Person"})
#formatting my money fields to add the $ sign 
analysed_gender["Average Purchase Price"] = analysed_gender["Average Purchase Price"].map("${:.2f}".format)
analysed_gender["Total Purchase Value"] = analysed_gender["Total Purchase Value"].map("${:,.2f}".format)
analysed_gender["Avg Total Purchase per Person"] = analysed_gender["Avg Total Purchase per Person"].map("${:.2f}".format)

analysed_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#removing duplicates from the data 
Serial_no = purchase_data[["SN","Age","Gender"]]
serial_duplicate = Serial_no.drop_duplicates()

#cutting Age data ino bins and creating labes for bins
bins =  [0, 9, 14, 19, 24, 29, 34, 39, 1000000]
group_name = ["<10","10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
serial_duplicate["Age Sort"] = pd.cut(serial_duplicate["Age"], bins, labels = group_name)

#grouping data by the ages calculated
groupage = serial_duplicate[["Age Sort","SN", "Age"]]
groupagepercent= groupage.groupby(['Age Sort']).count()
 
#calculating the percentage age 
age_percent = (round(((groupagepercent["Age"]/gender_sum) * 100), 2))
groupagepercent["Percentage Of Players"] = age_percent.astype(str) + "%"

#refining and printing my summary date frame
groupagepercent = groupagepercent[["SN","Percentage Of Players"]]
groupagepercent = groupagepercent.rename(columns = {"SN":"Total Count"})
groupagepercent
```

* Optional: give the displayed data cleaner formatting

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#renaming columns
purchase_data = purchase_data.rename(columns={'SN':"Serial"})
sn_df = purchase_data.groupby(['Serial'])

#calulating mean, total purchase
meansn_price =  round(sn_df["Price"].mean(), 2)
totalsn_purchase = sn_df["Purchase ID"].count() * meansn_price
totalsn_purchase = pd.DataFrame(totalsn_purchase)

#merging results into a summary data frame
mergerprice = pd.merge(sn_df["Purchase ID"].count(), meansn_price, on ="Serial")
mergerprice = pd.merge(mergerprice, totalsn_purchase, on ="Serial")


mergerprice = mergerprice.rename(columns={'Purchase ID':"Purchase Count", 'Price':"Average Purchase Price", 
                                          0:"Total Purchase Value"})

#sorting  df by decsning Total purchase value
mergerprice = mergerprice.sort_values(["Total Purchase Value"], ascending=False)

#formatting currency fields to add the $ sign 
mergerprice["Average Purchase Price"] = mergerprice["Average Purchase Price"].map("${:.2f}".format)
mergerprice["Total Purchase Value"] = mergerprice["Total Purchase Value"].map("${:.2f}".format)


mergerprice.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#sorting data and getting rewuired fields 
items = purchase_data[["Item ID", "Item Name", "Price"]]

#multigrouping data 
groupeditems = items.groupby(["Item ID", "Item Name"])

#calsulating values needed for df
item_count = groupeditems["Price"].count()
item_total= groupeditems["Price"].sum()
item_price = item_total/item_count

#merging df 
merge_items = pd.merge(item_count, item_price, on =["Item ID","Item Name"])

#Reanming Df columns
items_sorted = merge_items.rename(columns= {"Price_x":"Purchase Count","Price_y":"Price","Price":"Item Price"})
items_sorted = items_sorted.rename(columns= {"Price":"Items Price"})

items_sorted["Total Purchase Value"]=  round(items_sorted["Purchase Count"] * items_sorted["Items Price"], 2)

#sorting  df by descending Purchase count
top_purchase_count = items_sorted.sort_values("Purchase Count", ascending = False)

#formatting currency fields to add the $ sign 
top_purchase_count["Items Price"] = top_purchase_count["Items Price"].map("${:.2f}".format)
top_purchase_count["Total Purchase Value"] = top_purchase_count["Total Purchase Value"].map("${:.2f}".format)


top_purchase_count.head()
```
